## Apply clustering algorithms


- Apply K-Means, Mean-Shift, and Affinity Propagation algorithms to the preprocessed dataset.

- Import the necessary clustering algorithms from scikit-learn.


```python
# Import clustering algorithms ðŸ‡¬ðŸ‡§ Importar algoritmos de clustering ðŸ‡§ðŸ‡·
from sklearn.cluster import KMeans, MeanShift, AffinityPropagation, estimate_bandwidth
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score
```

Preprocessing Data

```python
# Standardize features ðŸ‡¬ðŸ‡§ Padronizar caracterÃ­sticas ðŸ‡§ðŸ‡·
scaler = StandardScaler()
df = pd.DataFrame(scaler.fit_transform(df[['Coluna1', 'Coluna2']]), columns=['Coluna1', 'Coluna2'])
df
```

**Reasoning**:
Apply the KMeans algorithm to the preprocessed data and store the cluster labels.



```python
inertia = []
silhouette = []
K = []

for k in range(2, 11):
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10) # Explicitly set n_init
    kmeans_labels = kmeans.fit(df)
    inertia.append(kmeans_labels.inertia_)
    silhouette.append(silhouette_score(df, kmeans_labels.labels_))
    K.append(k)

plt.figure(figsize=(12, 6))
sns.lineplot(x=range(2, 11), y=inertia, marker='o')
plt.title('KMeans Inertia vs Number of Clusters - KMeans InÃ©rcia vs NÃºmero de Clusters')
plt.xlabel('Number of Clusters (K) - NÃºmero de Clusters (K)')
plt.ylabel('Inertia - InÃ©rcia')
plt.show()

pd.DataFrame({'K': K, 'Silhouette': silhouette})
```

Since the Elbow Method did not indicate a clear optimal K, we selected K = 7 because the Silhouette score confirms this choice (Silhouette score = 0.6488).

Como o MÃ©todo do Cotovelo (Elbow Method) nÃ£o indicou um K Ã³timo claramente definido, selecionamos K = 7 porque a pontuaÃ§Ã£o de Silhouette confirma essa escolha (pontuaÃ§Ã£o de Silhouette â‰ˆ 0,6546).

```python
# Apply KMeans clustering ðŸ‡¬ðŸ‡§ Aplicar clustering KMeans ðŸ‡§ðŸ‡·
kmeans = KMeans(n_clusters=7, random_state=42, n_init=10) # Explicitly set n_init
df['kmeans_labels'] = kmeans.fit_predict(df)

# Display the first few rows with new labels ðŸ‡¬ðŸ‡§ Exibir as primeiras linhas com novos rÃ³tulos ðŸ‡§ðŸ‡·
print("\nðŸ‡¬ðŸ‡§ First few rows with KMeans labels: ðŸ‡§ðŸ‡· Primeiras linhas com rÃ³tulos KMeans:")
display(df.head())
```


Apply the Mean-Shift algorithm to the preprocessed data and store the cluster labels.



```python
# Estimar banda no espaÃ§o de atributos apenas ðŸ‡¬ðŸ‡§/ðŸ‡§ðŸ‡·
tam_banda = estimate_bandwidth(df[['Coluna1', 'Coluna2']].values, quantile=0.25, n_samples=df.shape[0], random_state=42)
print(f"Banda estimada: {tam_banda:.4f}")


# Apply Mean-Shift clustering ðŸ‡¬ðŸ‡§ Aplicar clustering Mean-Shift ðŸ‡§ðŸ‡·
meanshift = MeanShift(bandwidth = tam_banda, bin_seeding=True)
df['meanshift_labels'] = meanshift.fit_predict(df[['Coluna1', 'Coluna2']].values) # Fit only on the data columns


n_clusters_ms = len(df['meanshift_labels'].unique())
print(f"NÂº de clusters (MeanShift): {n_clusters_ms}")

# Display the first few rows with new labels ðŸ‡¬ðŸ‡§ Exibir as primeiras linhas com novos rÃ³tulos ðŸ‡§ðŸ‡·
print("\nðŸ‡¬ðŸ‡§ First few rows with Mean-Shift labels: ðŸ‡§ðŸ‡· Primeiras linhas com rÃ³tulos Mean-Shift:")
display(df.head())
```


Apply the Affinity Propagation algorithm to the preprocessed data and store the cluster labels.



```python
# Apply Affinity Propagation clustering ðŸ‡¬ðŸ‡§ Aplicar clustering Affinity Propagation ðŸ‡§ðŸ‡·
dfs = []
affinity_propagations = []
for preference in [None, -50, -40, -30, -20, -10, -5]:
    affinity_propagation = AffinityPropagation(
        damping=0.9,           # Prevents numerical oscillations
        max_iter=200,          # Maximum iterations
        convergence_iter=15,   # Iterations without change to declare convergence
        random_state=42,        # For reproducibility
        preference=preference   # Set preference value
    )
    df['affinity_propagation_labels'] = affinity_propagation.fit_predict(df[['Coluna1', 'Coluna2']].values) # Fit only on the data columns
    dfs.append(df.copy())
    affinity_propagations.append(affinity_propagation)

    # Get number of clusters and exemplars ðŸ‡¬ðŸ‡§ Obter nÃºmero de clusters e exemplares ðŸ‡§ðŸ‡·
    print(f'Preference: {preference} - Number of clusters (Affinity Propagation): {len(affinity_propagation.cluster_centers_)}')
    print(f'Silhouette Score (Affinity Propagation): {silhouette_score(df, affinity_propagation.labels_):.4f}')
    # Display the first few rows with new labels ðŸ‡¬ðŸ‡§ Exibir as primeiras linhas com novos rÃ³tulos ðŸ‡§ðŸ‡·
    '''print("\nðŸ‡¬ðŸ‡§ First few rows with Affinity Propagation labels: ðŸ‡§ðŸ‡· Primeiras linhas com rÃ³tulos Affinity Propagation:")
    display(df.head())'''
```

```python
# Apply Affinity Propagation clustering ðŸ‡¬ðŸ‡§ Aplicar clustering Affinity Propagation ðŸ‡§ðŸ‡·
affinity_propagation = AffinityPropagation(
    damping=0.9,           # Prevents numerical oscillations
    max_iter=200,          # Maximum iterations
    convergence_iter=15,   # Iterations without change to declare convergence
    random_state=42        # For reproducibility
)
df['affinity_propagation_labels'] = affinity_propagation.fit_predict(df[['Coluna1', 'Coluna2']].values) # Fit only on the data columns

# Get number of clusters and exemplars ðŸ‡¬ðŸ‡§ Obter nÃºmero de clusters e exemplares ðŸ‡§ðŸ‡·
print(f'Silhouette Score (Affinity Propagation): {silhouette_score(df, affinity_propagation.labels_):.4f}')
# Display the first few rows with new labels ðŸ‡¬ðŸ‡§ Exibir as primeiras linhas com novos rÃ³tulos ðŸ‡§ðŸ‡·
print("\nðŸ‡¬ðŸ‡§ First few rows with Affinity Propagation labels: ðŸ‡§ðŸ‡· Primeiras linhas com rÃ³tulos Affinity Propagation:")
display(df.head())
```


Import plotting libraries and set the plotting style and color palette for dark mode with turquoise tones.




Create a figure and subplots, then plot the results of each clustering algorithm on its respective subplot with titles and a main title.



```python
fig, (up, middle, down) = plt.subplots(3, 1, figsize=(12, 26))

# Plot KMeans clustering results in a separate figure ðŸ‡¬ðŸ‡§ Plotar resultados do clustering KMeans em uma figura separada ðŸ‡§ðŸ‡·
#plt.figure(figsize=(12, 8))
sns.scatterplot(x='Coluna1', y='Coluna2', hue='kmeans_labels', data=df, palette='viridis', ax = up)
sns.scatterplot(x = kmeans.cluster_centers_[:,0], y = kmeans.cluster_centers_[:, 1], s = 150, color ='black', marker='X', label = 'Centroides', ax = up)
up.set_title('K-Means Clustering / Agrupamento K-Means')
up.set_xlabel('Coluna1')
up.set_ylabel('Coluna2')
fig.tight_layout(pad=1.2)


# Plot Mean-Shift clustering results in a separate figure ðŸ‡¬ðŸ‡§ Plotar resultados do clustering Mean-Shift em uma figura separada ðŸ‡§ðŸ‡·
#plt.figure(figsize=(12, 8))
sns.scatterplot(x='Coluna1', y='Coluna2', hue='meanshift_labels', data=df, palette='viridis', ax = middle)
sns.scatterplot(x = meanshift.cluster_centers_[:,0], y = meanshift.cluster_centers_[:, 1], s = 150, color ='black', marker='X', label = 'Centroides', ax = middle)
middle.set_title('Mean-Shift Clustering / Agrupamento Mean-Shift')
middle.set_xlabel('Coluna1')
middle.set_ylabel('Coluna2')
fig.tight_layout(pad=1.2)


# Plot Affinity Propagation clustering results in a separate figure ðŸ‡¬ðŸ‡§ Plotar resultados do clustering Affinity Propagation em uma figura separada ðŸ‡§ðŸ‡·
#plt.figure(figsize=(12, 8))
palette = sns.color_palette("husl", 46)
sns.scatterplot(x='Coluna1', y='Coluna2', hue='affinity_propagation_labels', data=df, palette=palette, legend = False, ax = down)
sns.scatterplot(x = affinity_propagation.cluster_centers_[:,0], y = affinity_propagation.cluster_centers_[:, 1], s = 150, color ='black', marker='X', label = 'Centroides', ax = down)
down.set_title('Affinity Propagation Clustering / Agrupamento por PropagaÃ§Ã£o de Afinidade')
down.set_xlabel('Coluna1')
down.set_ylabel('Coluna2')
fig.tight_layout(pad=1.2)
plt.show()
```
