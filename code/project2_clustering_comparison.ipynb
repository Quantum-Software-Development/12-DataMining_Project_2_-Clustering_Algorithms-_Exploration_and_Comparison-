Import pandas and load the dataset into a DataFrame.



```python
# Import pandas library ðŸ‡¬ðŸ‡§ Importar a biblioteca pandas ðŸ‡§ðŸ‡·
import pandas as pd

# Load the dataset ðŸ‡¬ðŸ‡§ Carregar o dataset ðŸ‡§ðŸ‡·
df = pd.read_csv('/content/Grupo4.csv')

# Display the first few rows of the DataFrame ðŸ‡¬ðŸ‡§ Exibir as primeiras linhas do DataFrame ðŸ‡§ðŸ‡·
display(df.head())
```


Get the number of rows and columns and display basic statistical measures.



```python
# Get the number of rows and columns ðŸ‡¬ðŸ‡§ Obter o nÃºmero de linhas e colunas ðŸ‡§ðŸ‡·
num_rows, num_cols = df.shape

# Print the number of rows and columns ðŸ‡¬ðŸ‡§ Imprimir o nÃºmero de linhas e colunas ðŸ‡§ðŸ‡·
print(f"ðŸ‡¬ðŸ‡§ Number of rows: {num_rows}, Number of columns: {num_cols}")
print(f"ðŸ‡§ðŸ‡· NÃºmero de linhas: {num_rows}, NÃºmero de colunas: {num_cols}")

# Display descriptive statistics ðŸ‡¬ðŸ‡§ Exibir estatÃ­sticas descritivas ðŸ‡§ðŸ‡·
display(df.describe())
```


The first step is to remove the 'Unnamed: 0' column if it exists, followed by handling missing values and removing duplicates, as per the subtask instructions.



```python
# Check if 'Unnamed: 0' column exists and drop it if it does ðŸ‡¬ðŸ‡§ Verificar se a coluna 'Unnamed: 0' existe e removÃª-la se sim ðŸ‡§ðŸ‡·
if 'Unnamed: 0' in df.columns:
    df.drop('Unnamed: 0', axis=1, inplace=True)
    print("ðŸ‡¬ðŸ‡§ 'Unnamed: 0' column dropped. ðŸ‡§ðŸ‡· Coluna 'Unnamed: 0' removida.")
else:
    print("ðŸ‡¬ðŸ‡§ 'Unnamed: 0' column not found. ðŸ‡§ðŸ‡· Coluna 'Unnamed: 0' nÃ£o encontrada.")

# Calculate the median of each column ðŸ‡¬ðŸ‡§ Calcular a mediana de cada coluna ðŸ‡§ðŸ‡·
column_medians = df.median()
print("\nðŸ‡¬ðŸ‡§ Medians of each column: ðŸ‡§ðŸ‡· Medianas de cada coluna:")
display(column_medians)

# Fill missing values with the calculated medians ðŸ‡¬ðŸ‡§ Preencher valores faltantes com as medianas calculadas ðŸ‡§ðŸ‡·
df.fillna(column_medians, inplace=True)
print("\nðŸ‡¬ðŸ‡§ Missing values filled with medians. ðŸ‡§ðŸ‡· Valores faltantes preenchidos com as medianas.")

# Check for and remove duplicate rows ðŸ‡¬ðŸ‡§ Verificar e remover linhas duplicadas ðŸ‡§ðŸ‡·
initial_rows = df.shape[0]
df.drop_duplicates(inplace=True)
rows_after_duplicates = df.shape[0]
print(f"\nðŸ‡¬ðŸ‡§ Number of duplicate rows removed: {initial_rows - rows_after_duplicates} ðŸ‡§ðŸ‡· NÃºmero de linhas duplicadas removidas: {initial_rows - rows_after_duplicates}")


# Display the first few rows of the preprocessed DataFrame ðŸ‡¬ðŸ‡§ Exibir as primeiras linhas do DataFrame prÃ©-processado ðŸ‡§ðŸ‡·
print("\nðŸ‡¬ðŸ‡§ First few rows of the preprocessed DataFrame: ðŸ‡§ðŸ‡· Primeiras linhas do DataFrame prÃ©-processado:")
display(df.head())

# Print the number of remaining rows and columns ðŸ‡¬ðŸ‡§ Imprimir o nÃºmero de linhas e colunas restantes ðŸ‡§ðŸ‡·
num_rows_preprocessed, num_cols_preprocessed = df.shape
print(f"\nðŸ‡¬ðŸ‡§ Number of rows after preprocessing: {num_rows_preprocessed}, Number of columns after preprocessing: {num_cols_preprocessed}")
print(f"ðŸ‡§ðŸ‡· NÃºmero de linhas apÃ³s o prÃ©-processamento: {num_rows_preprocessed}, NÃºmero de colunas apÃ³s o prÃ©-processamento: {num_cols_preprocessed}")
```

## Apply clustering algorithms


- Apply K-Means, Mean-Shift, and Affinity Propagation algorithms to the preprocessed dataset.

- Import the necessary clustering algorithms from scikit-learn.


```python
# Import clustering algorithms ðŸ‡¬ðŸ‡§ Importar algoritmos de clustering ðŸ‡§ðŸ‡·
from sklearn.cluster import KMeans, MeanShift, AffinityPropagation
```

**Reasoning**:
Apply the KMeans algorithm to the preprocessed data and store the cluster labels.



```python
# Apply KMeans clustering ðŸ‡¬ðŸ‡§ Aplicar clustering KMeans ðŸ‡§ðŸ‡·
kmeans = KMeans(n_clusters=3, random_state=42, n_init=10) # Explicitly set n_init
df['kmeans_labels'] = kmeans.fit_predict(df)

# Display the first few rows with new labels ðŸ‡¬ðŸ‡§ Exibir as primeiras linhas com novos rÃ³tulos ðŸ‡§ðŸ‡·
print("\nðŸ‡¬ðŸ‡§ First few rows with KMeans labels: ðŸ‡§ðŸ‡· Primeiras linhas com rÃ³tulos KMeans:")
display(df.head())
```


Apply the Mean-Shift algorithm to the preprocessed data and store the cluster labels.



```python
# Apply Mean-Shift clustering ðŸ‡¬ðŸ‡§ Aplicar clustering Mean-Shift ðŸ‡§ðŸ‡·
meanshift = MeanShift()
df['meanshift_labels'] = meanshift.fit_predict(df[['Coluna1', 'Coluna2']]) # Fit only on the data columns

# Display the first few rows with new labels ðŸ‡¬ðŸ‡§ Exibir as primeiras linhas com novos rÃ³tulos ðŸ‡§ðŸ‡·
print("\nðŸ‡¬ðŸ‡§ First few rows with Mean-Shift labels: ðŸ‡§ðŸ‡· Primeiras linhas com rÃ³tulos Mean-Shift:")
display(df.head())
```


Apply the Affinity Propagation algorithm to the preprocessed data and store the cluster labels.



```python
# Apply Affinity Propagation clustering ðŸ‡¬ðŸ‡§ Aplicar clustering Affinity Propagation ðŸ‡§ðŸ‡·
affinity_propagation = AffinityPropagation()
df['affinity_propagation_labels'] = affinity_propagation.fit_predict(df[['Coluna1', 'Coluna2']]) # Fit only on the data columns

# Display the first few rows with new labels ðŸ‡¬ðŸ‡§ Exibir as primeiras linhas com novos rÃ³tulos ðŸ‡§ðŸ‡·
print("\nðŸ‡¬ðŸ‡§ First few rows with Affinity Propagation labels: ðŸ‡§ðŸ‡· Primeiras linhas com rÃ³tulos Affinity Propagation:")
display(df.head())
```


Import plotting libraries and set the plotting style and color palette for dark mode with turquoise tones.



```python
# Import plotting libraries ðŸ‡¬ðŸ‡§ Importar bibliotecas de plotagem ðŸ‡§ðŸ‡·
import matplotlib.pyplot as plt
import seaborn as sns

# Set plotting style to darkgrid ðŸ‡¬ðŸ‡§ Definir estilo de plotagem para darkgrid ðŸ‡§ðŸ‡·
sns.set_style('darkgrid')

# Set color palette to a turquoise-toned palette ðŸ‡¬ðŸ‡§ Definir paleta de cores para um tom turquesa ðŸ‡§ðŸ‡·
sns.set_palette('viridis')
```


Create a figure and subplots, then plot the results of each clustering algorithm on its respective subplot with titles and a main title.



```python
# Plot KMeans clustering results in a separate figure ðŸ‡¬ðŸ‡§ Plotar resultados do clustering KMeans em uma figura separada ðŸ‡§ðŸ‡·
plt.figure(figsize=(12, 8))
sns.scatterplot(x='Coluna1', y='Coluna2', hue='kmeans_labels', data=df, palette='viridis', legend=False)
plt.title('K-Means Clustering / Agrupamento K-Means')
plt.xlabel('Coluna1')
plt.ylabel('Coluna2')
plt.tight_layout()
plt.show()

# Plot Mean-Shift clustering results in a separate figure ðŸ‡¬ðŸ‡§ Plotar resultados do clustering Mean-Shift em uma figura separada ðŸ‡§ðŸ‡·
plt.figure(figsize=(12, 8))
sns.scatterplot(x='Coluna1', y='Coluna2', hue='meanshift_labels', data=df, palette='viridis', legend=False)
plt.title('Mean-Shift Clustering / Agrupamento Mean-Shift')
plt.xlabel('Coluna1')
plt.ylabel('Coluna2')
plt.tight_layout()
plt.show()

# Plot Affinity Propagation clustering results in a separate figure ðŸ‡¬ðŸ‡§ Plotar resultados do clustering Affinity Propagation em uma figura separada ðŸ‡§ðŸ‡·
plt.figure(figsize=(12, 8))
sns.scatterplot(x='Coluna1', y='Coluna2', hue='affinity_propagation_labels', data=df, palette='viridis', legend=False)
plt.title('Affinity Propagation Clustering / Agrupamento por PropagaÃ§Ã£o de Afinidade')
plt.xlabel('Coluna1')
plt.ylabel('Coluna2')
plt.tight_layout()
plt.show()
```
